```python
# prompt: cpu value count

data['Cpu'].value_counts()
```

```python
data['cpuName'] = data['Cpu'].apply(lambda x:' '.join(x.split()[0:3]))
```

```python
data.head()
```

```python
data['cpuName'].value_counts()
```

```python
def setProcessor(name):
  if name == 'Intel Core i7' or name =='Intel Core i5' or name == 'Intel Core i3':
    return name
  else:
    if name.split()[0]=='AMD':
      return 'AMD'
    else:
      return 'other'
```

```python
data['cpuName'] = data['cpuName'].apply(setProcessor)
```

```python
data['cpuName'].value_counts()
```

```python
data['gpuName'] = data['Gpu'].apply(lambda x:"".join(x.split()[0:1]))
```

```python
data = data[data['gpuName'] != 'ARM']
```

```python
data['OpSys'].value_counts()
```

```python
def setOS(input):
  if input == 'Windows 10' or input == 'Windows 7' or input == 'Windows 10 S':
    return 'windows'
  elif input == 'Mac OS X' or input == 'macOS':
    return 'mac'
  elif input == 'Linux':
    return 'Linux'
  else:
    return 'other'
```

```python
data['OpSys'] = data['OpSys'].apply(setOS)
```

```python
data['OpSys'].value_counts()
```
