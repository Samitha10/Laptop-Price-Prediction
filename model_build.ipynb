```python
# prompt: cpu value count

data['Cpu'].value_counts()
```

```python
data['cpuName'] = data['Cpu'].apply(lambda x:' '.join(x.split()[0:3]))
```

```python
data.head()
```

```python
data['cpuName'].value_counts()
```

```python
def setProcessor(name):
  if name == 'Intel Core i7' or name =='Intel Core i5' or name == 'Intel Core i3':
    return name
  else:
    if name.split()[0]=='AMD':
      return 'AMD'
    else:
      return 'other'
```

```python
data['cpuName'] = data['cpuName'].apply(setProcessor)
```

```python
data['cpuName'].value_counts()
```

```python
data
```

```python
data['Gpu'].value_counts()
```

```python
data['gpuName'] = data['Gpu'].apply(lambda x:"".join(x.split()[0:1]))
```

```python
data['gpuName'].value_counts()
```

```python
data = data[data['gpuName'] != 'ARM']
```

```python
data['gpuName'].value_counts()
```

```python
data.shape
```

```python
data['OpSys'].value_counts()
```

```python
def setOS(input):
  if input == 'Windows 10' or input == 'Windows 7' or input == 'Windows 10 S':
    return 'windows'
  elif input == 'Mac OS X' or input == 'macOS':
    return 'mac'
  elif input == 'Linux':
    return 'Linux'
  else:
    return 'other'
```

```python
data['OpSys'] = data['OpSys'].apply(setOS)
```

```python
data['OpSys'].value_counts()
```

```python
# prompt: delete those columns laptop_ID Inches Product Screenresolution Cpu Gpu

data = data.drop(['laptop_ID', 'Inches', 'Product', 'ScreenResolution', 'Cpu', 'Gpu'], axis=1)

```

```python
data
```

```python
data = pd.get_dummies(data)
data
```

```python
data.shape
```

```python
# prompt: export  'data' to exel file

#data.to_excel('cleaned_data.xlsx', index=False)
```

```python
X = data.drop('Price_euros',axis=1)
y = data['Price_euros']
```

```python
import sklearn
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python
X_train.shape
```

```python
def modelAcc(model):
  model.fit(X_train,y_train)
  acc = model.score(X_test,y_test)
  print(str(acc))
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
modelAcc(lr)
```

```python
from sklearn.linear_model import Lasso
lasso = Lasso()
modelAcc(lasso)
```

```python
from sklearn.tree import DecisionTreeRegressor
dtr = DecisionTreeRegressor()
modelAcc(dtr)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf = RandomForestRegressor()
modelAcc(rf)
```

```python
from sklearn.model_selection import GridSearchCV
parameters = {
    'n_estimators':[10,50,100],
    'criterion':['squared_error','absolute_error','poisson']
}
grid_obj = GridSearchCV(estimator=rf,param_grid=parameters)
grid_fit = grid_obj.fit(X_train,y_train)
best_model = grid_fit.best_estimator_
best_model
```

```python
best_model.score(X_test,y_test)
```

```python
import pickle
with open('predictor.pickle','wb') as file:
  pickle.dump(best_model,file)
```
